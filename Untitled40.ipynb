```python
# Importing nessecary libraries
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import decomposition
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
# Plotting the distribtions
sns.set(rc={"figure.figsize":(15, 5)})
g = sns.kdeplot(data = df)
g.set(title='All features distributions');
```

```python
# Checking for dtypes and missing or dublicated values
df.info()
```

```python
# Remove the dublicated values
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# Correlation the between multiple variables
corr = df.corr()
corr
```

```python
# Finding the maximum number of the clusters
limit = int((df.shape[0]//2)**0.5)
limit
```

```python
# Using the elbow method to determine the optimal number of clusters for k-means clustering
wcss = {}
 
for k in range(2,limit+1):
    model = KMeans(n_clusters=k)
    model.fit(df)
    wcss[k] = model.inertia_
```

```python
plt.plot(wcss.keys(), wcss.values(), 'gs-')
plt.xlabel('Values of "k"')
plt.title('The elbow method')
plt.ylabel('WCSS')
plt.show()
```

optimal number of clusters for k-means clustering is 3

```python
# Using "Principal Component Analysis" PCA to reduce the dimensionality to 2 dimensions
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=2)
X_centered = X - X.mean(axis=0)
pca.fit(X_centered)
X_pca = pca.transform(X_centered)
```

```python
# Plotting the results of PCA
plt.plot(X_pca[y == 0, 0], X_pca[y == 0, 1], 'bo', label='Setosa')
plt.plot(X_pca[y == 1, 0], X_pca[y == 1, 1], 'go', label='Versicolour')
plt.plot(X_pca[y == 2, 0], X_pca[y == 2, 1], 'ro', label='Virginica')
plt.legend(loc=0);
```

```python
# Applying kmeans to the dataset (Creating the kmeans classifier)
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Plotting the centroids of the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend().remove()
```
